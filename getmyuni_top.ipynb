```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
#import pandas inorder to create dataframes
```

```python
url="https://www.getmyuni.com/all-colleges"
```

```python
# Define headers to mimic a web browserhttps://www.makemytrip.com/hotels/hotel-listing/?checkin=08052024&checkout=08062024&locusId=CTMAA&locusType=city&city=CTMAA&country=IN&searchText=Chennai&roomStayQualifier=2e0e&_uCurrency=INR&reference=hotel&type=city&rsc=1e2e0e
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36",
    "Referer": "https://www.getmyuni.com/all-colleges"
}

try:
    # send a Get request with the defined headers
    response = requests.get(url, headers=headers)
   
    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        print("Access Granted. Status Code:", response.status_code)
        # if you need to do something with the webpage content, you can access it using response.text
   
    else:
        print("unexpected Status Code:", response.status_code)


except requests.exceptions.requestException as e:
    print("Error:",e)
```

```python
soup=BeautifulSoup(response.content,'html.parser')
soup
```

```python
names=soup.find_all('h2',class_='college__name')
city=soup.find_all('div',class_='detail__list__mobile')
fees=soup.find_all('span',class_='highlight__value')
```

```python
college_names = []
college_cities = []
college_fees = []
```

```python
for name, city, fee in zip(names, city, fees):
    college_names.append(name.text.strip())
    college_cities.append(city.text.strip())
    college_fees.append(fee.text.strip())
```

```python
college_data = {'college_names':college_names, 'college_cities':college_cities, 'college_fees':college_fees}
print(college_data)
```

```python
model=pd.DataFrame(data=college_data)
model.to_csv("college_data.csv")
model
```
